# Synthetic DGP Walkthrough

Validate uplift, bandit, and DAG data generators with known ground truth.

```python
import sys
sys.path.insert(0, '../src')

import matplotlib.pyplot as plt
import numpy as np

from causal_ml.data.uplift_dgp import UpliftDGPConfig, generate_uplift_data
from causal_ml.data.bandit_dgp import BanditDGPConfig, generate_bandit_data
from causal_ml.data.dag_dgp import DAGDGPConfig, generate_dag_data
```

```python
uplift = generate_uplift_data(UpliftDGPConfig(n_samples=2000, confounding_strength=1.0, seed=42))
tau = uplift.ground_truth['tau']
e = uplift.ground_truth['propensity']

fig, axes = plt.subplots(1, 2, figsize=(10, 4))
axes[0].hist(tau, bins=40, edgecolor='k')
axes[0].set_title('Heterogeneous treatment effects tau(x)')
axes[1].hist(e, bins=40, edgecolor='k')
axes[1].set_title('Propensity scores e(x)')
plt.tight_layout()
plt.show()
```

```python
bandit = generate_bandit_data(BanditDGPConfig(n_samples=2000, seed=42))
print('Oracle value:', bandit.ground_truth['oracle_value'])
print('Behavior value:', bandit.ground_truth['behavior_value'])
```

```python
dag = generate_dag_data(DAGDGPConfig(n_samples=1000, n_nodes=5, seed=42))
print('Adjacency matrix:')
print(dag.ground_truth['adjacency'])
```
